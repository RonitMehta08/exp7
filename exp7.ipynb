While Loop

```python
#print number till 5
i = 1
while(i<=5):
  print(i)
  i +=1
```

```python
#print number till n numbers
n = int(input("Enter N:"))
i = 1
while(i<=n):
  print(i)
  i += 1
```

```python
#factorial of a number
n = int(input("Enter a Number: "))
fact = 1

while n>0:
  fact *= n
  n -= 1

print("Factorial:", fact)
```

```python
#fibbonacci series using while loop

n = int(input("Enter a Number: "))

a = 0
b = 1
i = 1
while i <= n:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
limit = int(input("Enter Limit: "))
a = 0
b = 1
while a <= limit:
  print(a, end = " ")
  a, b = b, a+b
```

```python
#Reverse a number
num = int(input("Enter a number: "))
rev = 0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
print("Reversed: ", rev)
```

```python
#check pallindrome number/string

num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0:
  rev = rev*10 + num%10
  num //= 10

if temp == rev:
  print("Pallindrome")
else:
  print("Not Pallindrome")
```

```python
s = input("Enter a string: ")
i, j = 0, len(s)-1
is_pallindrom = True

while i < j:
  if s[i] != s[j]:
    is_pallindrom = False
    break
  i += 1
  j -= 1

if is_pallindrom:
  print("Pallindrome")
else:
  print("Not Pallindrome")
```

```python
st = input("Enter a string: ")
rev = st[::-1]
if st == rev:
  print("Pallindrome")
else:
  print("Not Pallindrome")
```

```python
#count digits in a number

n = int(input("Enter a Number: "))
count = 0
while n > 0:
  n //= 10
  count += 1

print("Number of digits:", count)
```

```python
# Exit the loop when i is 3
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
# Search an element in a list
nums = [10, 20,30,40,50]
key = int(input("Enter an element to search in the list: "))
i = 0
while i < len(nums):
    if nums[i] == key:
      print("Element found at index", i+1)
      break
    i += 1
else:
  print("Element not found in the list")
```
